#Analysis of CD56 and Perforin Expression in a CyTOF Dataset

The purpose of this analysis was to examine the distributions of CD56 and Perforin, two immunologically significant markers, within a CyTOF dataset. CyTOF (Cytometry by Time-Of-Flight) is an advanced flow cytometry technology enabling the measurement of multiple parameters at the single-cell level.

The markers chosen (CD56 and Perforin) for this study are pivotal to the functional characterization of natural killer (NK) cells.

#Importing Libraries and Loading Dataset


```python
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
df = pd.read_csv('cytof_one_experiment.csv')
```

#Visualizations

```python
# Set the aesthetic style of the plots
sns.set_style("whitegrid")
```

##CD56 and Perforin Histograms and Density Estimates

```python
# Distribution of CD56
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['CD56'], kde=True, color='skyblue', bins=30)
plt.title('Histogram of CD56')

plt.subplot(1, 2, 2)
sns.kdeplot(df['CD56'], fill=True, color='olive')
plt.title('Density Estimate of CD56')
plt.show()

# Distribution of Perforin
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['Perforin'], kde=True, color='tomato', bins=30)
plt.title('Histogram of Perforin')

plt.subplot(1, 2, 2)
sns.kdeplot(df['Perforin'], fill=True, color='gold')
plt.title('Density Estimate of Perforin')
plt.show()
```

The distribution of both CD56 and Perforin was assessed using histograms and kernel density estimates (KDE). These visualizations revealed that both markers exhibit right-skewed distributions, suggesting a variation in expression levels across the cell population, which is typical due to the heterogeneity of NK cell states.

##Q-Q Plots for CD56 and Perforin Distributions

```python
#Q-Q plot for CD56
qqplot_cd56 = sm.qqplot(df['CD56'], line='s')
plt.title('Q-Q Plot for CD56')
plt.show()
#Q-Q plot for Perforin
qqplot_perforin = sm.qqplot(df['Perforin'], line='s')
plt.title('Q-Q Plot for Perforin')
plt.show()
```

Q-Q plots were used to compare the quantiles of each marker's expression levels to those of a standard normal distribution. The plots for both CD56 and Perforin show that neither follows a normal distribution, with deviations indicating skewness and variability in expression levels.



#Conclusion


The observed skewness in their distributions reflects the biological variability inherent in the immune response. This shows the importance of these markers (such as the ones we used: CD56 and Perforin)in differentiating different NK cell states and functionalities.
